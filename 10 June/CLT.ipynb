### Central Limit Theorem 

##### It states that if you take enough random samples from any dataset, the averages of those samples will always form a perfent, symmetric bell curve no matter how chaotic or wird the original dataset looks

##### Example : If you want to know the average income of 1.4 billion people, you don't have to survey all of them. By surveying a few random sample groups of 100 people, the CLT mathematically guarantees that the average of your samples will point directly toward the real, hidden national average.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
np.random.seed(42)
raw_data = np.random.exponential(scale=2, size = 10000)
sample_mean = []

for _ in range(10000):
    random_sample = np.random.choice(raw_data, size = 100)
    sample_mean.append(np.mean(random_sample))

plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
sns.histplot(raw_data, kde=True, color='red')
plt.title("Raw data")

plt.subplot(1, 2, 2)
sns.histplot(sample_mean, kde=True, color='blue')
plt.title("Perfect curve due to CLT")

plt.tight_layout()
plt.show()
```
