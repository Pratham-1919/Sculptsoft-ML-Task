### Dataframe in pyython 
In Python, a DataFrame is a two-dimensional, spreadsheet-like data structure where data is organized in a tabular format using rows and columns

```python
import pandas as pd
import numpy as np


raw_data = {
    'Transaction_ID': [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1001, 1008, 1009, 1010],
    'Bedrooms': [3, 4, 2, np.nan, 5, 3, 4, 3, 2, 3, np.nan],
    'Square_Feet': [1500, 2200, np.nan, 1200, 3200, 1800, 2400, 1500, 950, 1650, 1100],
    'Neighborhood': ['Suburb', 'City', 'Suburb', 'Rural', 'City', np.nan, 'Rural', 'Suburb', 'Rural', 'City', 'Suburb'],
    'Has_Pool': ['Yes', 'No', 'No', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No', 'No'],
    'House_Price': [250000, 420000, 180000, 150000, 550000, 290000, 380000, 250000, 130000, 310000, 165000]
}

df = pd.DataFrame(raw_data)
print(df)
```

```python
df.dtypes
```

```python
df["bathroom"] = df["Bedrooms"]-1
```

```python
df
```

```python
df1 = df
```

```python
df1
```

```python
df2 = df1.drop('Neighborhood', axis=1)
df2
```

```python
df3 = df2.copy()
df3
```

```python
df3["Bedrooms"] = df2["Bedrooms"].fillna(df2['Bedrooms'].mean())
```

```python
df3
```

```python
df3['Bedrooms'] = df3['Bedrooms'].round(0).astype('Int64')
df3
```

```python
df4 = df3
df4["Square_Feet"] = df3['Square_Feet'].fillna(df3["Square_Feet"].mean()).round(0).astype("Int64")
df4
```

```python
df5 = df4.copy()
df5['bathroom'] = df5['bathroom'].fillna(df5['bathroom'].median())
df5
```

```python
pool = df5.groupby("Has_Pool")
pool
```

```python
df5["Has_Pool"]
```

```python
pool_map = {"Yes" : 1, "No": 0}
df5["Has_Pool"] = df5["Has_Pool"].map(pool_map)
```
