### why pandas ?

In machine learning, models cannot read raw data formats like CSVs, text, or database tables directly. They require perfectly formatted, numerical matrices.

Pandas acts as the data bridge between raw information and machine learning algorithms. While NumPy handles the raw math, Pandas is used for Data Wrangling, Exploratory Data Analysis (EDA), and Feature Engineering.

### Series
A Series is a 1D (one-dimensional) array-like object. It is essentially a single column of data that can hold any data type

```python
import pandas as pd
import numpy as np
```

```python
ages = pd.Series([12,23,34], name = "Ages")
print(ages)
```
