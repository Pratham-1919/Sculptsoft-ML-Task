### Why Numpy 
###### In python we can use list but list is the scattered address of pointers which points every element in the list
###### It takes almost 28 bytes of space in the device
###### There is no vectorization in list if we are given 1000000 elements we need a loop
###### There can be different data type of each element which makes it time consuming
###### we need to write loops every time to preform mathematical operation 


#### Numpy is a continues block of memory allocation of same data type, a single interger in numpy takes 8 bytes of sapce in the deviuce
#### very easy to perform mathematically operations without loop
#### supports vectorization in array

### Advance numpy

```python
## Stacking vertical

a = np.array([1,2,3])
b = np.array([2,3,4])
c = np.vstack((a,b))
c
```

```python
## Stacking horizontal
a = np.array([1,2,3])
b = np.array([2,3,4])
c = np.hstack((a,b))
c
```

```python
## Splitting in matrix
matrix = np.array([
    [1,2,3,4],
    [2,3,4,5],
    [3,4,5,6],
    [4,5,6,7]
])
new_vmatix1, new_vmatix2 = np.vsplit(matrix, 2) 
print("Matrix 1:\n", new_vmatix1)
print("Matrix 2:\n", new_vmatix2)
```

```python
matrix = np.array([
    [1,2,3,4],
    [2,3,4,5],
    [3,4,5,6],
    [4,5,6,7]
])
new_vmatix1, new_vmatix2 = np.hsplit(matrix, 2) 
print("Matrix 1:\n", new_vmatix1)
print("Matrix 2:\n", new_vmatix2)
```

```python
## Linear Algebra using NumPy
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
result = A @ B
print(A.T)
result
```

```python
## solving a algebric equation
A = np.array([[1, 2], [3, 4]])
B = np.array([5, 11])
x = np.linalg.solve(A, B)
x
```
