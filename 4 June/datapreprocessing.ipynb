```python
import pandas as pd
```

```python

data1 = pd.DataFrame({
    'customer_id': [1, 2, 3],
    'name': ['John', 'Jane', 'Jack'],
    'age': [28, 34, 29]
})

data2 = pd.DataFrame({
    'customer_id': [1, 3, 4],
    'purchase_amount': [100.5, 85.3, 45.0],
    'purchase_date': ['2023-12-01', '2023-12-02', '2023-12-03']
})
    
merged_data = pd.merge(data1, data2, on='customer_id', how='inner')

print(merged_data)
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
data = pd.DataFrame({
    "country" :     ["India", "USA", "Canada", "India","USA"],
    "population" :  [10, 15, 10, 10, 15]
})

scaler = StandardScaler()
data["Scaled_coulmn"] = scaler.fit_transform(data[["population"]])


encoder = OneHotEncoder(sparse_output=False)
encoded_data = pd.DataFrame(encoder.fit_transform(data[["country"]]), columns=encoder.get_feature_names_out(['country']))

data = pd.concat([data, encoded_data], axis=1)

print(data)
```

```python
import numpy as np


data = pd.DataFrame({
    "Number" : [12,23,np.nan,67, 34, np.nan,65,45]
})
```

```python
print(data["Number"])
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy="mean")
data["Number"] = imputer.fit_transform(data[["Number"]])
print(data["Number"])
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

df = pd.DataFrame({'Sales': [100, 250, 400, 900, 10000]})

min_max = MinMaxScaler()
df['Sales_Normalized'] = min_max.fit_transform(df[['Sales']])

std_scaler = StandardScaler()
df['Sales_Standardized'] = std_scaler.fit_transform(df[['Sales']])

print(df)
```
