```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"E:\sculptsoft\1_June\EDA\pseudo_facebook.csv")
df.head()
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
print(df.isnull().sum())
```

```python
gender_map = {'male' : 1, 'female' : 0}
df["gender"] = df["gender"].map(gender_map).astype(int)
```

```python
df.head()
```

```python
print(df.isnull().sum())
```

```python
def rage_of_age(age):
    if 11<= age and age<=20:
        return '11-20'
    elif 21<= age and age<30:
        return '21-30'
    elif 31<=age and age<= 41:
        return '31-40'
    elif 41<=age and age<=50:
        return '41-50'
    elif 51<= age and age<=60:
        return '51-60'
    else:
        return '60+'
    
df['age_range'] = df['age'].map(rage_of_age)
df.head(100)
```

```python
group_counts = df["age_range"].value_counts().sort_index()
fig,ax = plt.subplots(figsize = (10,7))
ax.bar(group_counts.index.astype(str),group_counts.values)
plt.show()
```

```python
gender_counts = df["gender"].value_counts()
fig, ax = plt.subplots(figsize = (5,5))
ax.pie(
    gender_counts.values,
    labels= gender_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['skyblue', 'lightcoral'],
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)

ax.set_title("Gender Ratio Breakdown", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
