```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.normal(loc = 1500, scale = 300, size = 1000)
fig, ax = plt.subplots(figsize = (7,4))
ax.hist(data,bins= 30, color = "blue", alpha = 0.7)
ax.set_title("histplot")
ax.set_xlabel("value")
ax.set_ylabel('Frequency')
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()  
plt.show()
```

```python
x = np.random.normal(0,50,10)
y = x*2

plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Demo")
```

```python
plt.plot(x,y,'r--')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Demo")
# plt.plot(x,y,'b*-')
```

```python
data = [20,35,60,13,35,68,56,723,57,36]

fig,ax = plt.subplots(figsize = (5,5))
ax.boxplot(data)
plt.show()
```

```python
x = np.random.randint(0,10,10)
y = np.random.randint(0,10,10)

fig,ax = plt.subplots()
ax.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python

matrix = np.random.randint(10,101,size = (10,10))
print(matrix)

```

### ploting the chart of any share price

```python
year = [2021,2022,2023,2024,2025,2026]
ioc_price = [75.5,78.0,90.0,165.0,138.0,141.0]

fig, ax = plt.subplots(figsize = (8,5))
ax.plot(year,ioc_price,
        color='#00529B',
        linestyle='-',
        linewidth=2.5,
        marker='o',
        markersize=8,
        markerfacecolor='#FF6600',
        markeredgecolor='black',
        label='IOC Closing Price'
        )

ax.set_title('Indian Oil Corporation (IOC) Share Price History (2021 - 2026)', 
             fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Year', fontsize=11, fontweight='bold')
ax.set_ylabel('Share Price (INR)', fontsize=11, fontweight='bold')
ax.set_xticks(year)
ax.legend(loc='upper left')
ax.text(2023.8, 100, "2024 Bull Run\nRefining Margins Spike", 
        color='darkgreen', fontsize=9, fontweight='bold', bbox=dict(facecolor='white', alpha=0.6))
plt.tight_layout()
plt.show()
```
