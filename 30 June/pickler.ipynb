# Machine Learning Pipeline with Pickle Workflow

## Phase 1: Training Phase

```
                Load Dataset
                      │
                      ▼
          Data Preprocessing
 (Cleaning, Encoding, Scaling, Feature Engineering)
                      │
                      ▼
             Split the Dataset
            (Train/Test Split)
                      │
                      ▼
              Build Pipeline
 (Preprocessing + Machine Learning Model)
                      │
                      ▼
             Train the Pipeline
                pipe.fit(X_train, y_train)
                      │
                      ▼
         Evaluate the Model
      (Accuracy, F1, RMSE, etc.)
                      │
                      ▼
        Save the Trained Pipeline
       pickle.dump(pipe, "model.pkl")
```

---

## Phase 2: Deployment / Prediction Phase

```
          Load Saved Pipeline
        pickle.load("model.pkl")
                  │
                  ▼
        Receive New Input Data
                  │
                  ▼
 Pipeline Automatically Performs
      All Preprocessing Steps
 (Scaling, Encoding, Imputation, etc.)
                  │
                  ▼
      Trained Model Predicts
                  │
                  ▼
        Return Prediction
```

# Advantages of Using Pipeline + Pickle

- Train the model only once.
- Save the complete workflow in a single `.pkl` file.
- No need to manually perform preprocessing during prediction.
- Ensures the same preprocessing is applied during training and prediction.
- Easy to deploy in web applications (Flask, FastAPI, Django, Streamlit).
- Reduces the chances of preprocessing errors.
- Makes the model portable and reusable.
