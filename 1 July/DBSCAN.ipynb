### DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

DBSCAN is an unsupervised machine learning clustering algorithm that groups together points that are closely packed (high density) and marks points in low-density regions as noise (outliers).

Unlike K-Means and Hierarchical Clustering, DBSCAN does not require you to specify the number of clusters beforehand.

### Types of Points in DBSCAN

1. Core Point
 - **A point having at least min_samples neighbors inside the radius eps.**
2. Border Point
 - **A point that is close to a Core Point but does not itself have enough nearby neighbors to qualify as a Core Point.**
3. Noise Point
 - **A point that is not close enough to any Core Point.**

### When should we use DBSCAN?

    - **You don't know the number of clusters.**
    - **Your data contains outliers.**
    - **Clusters have irregular shapes.**
    - **Clusters are based on density rather than spherical shapes.**

### Major Parameters

    - **1. eps**
    - **2. min_samples**
    - **3. metric**
    

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
centers = [[1, 1], [-1, -1], [1, -1], [-1, 1]]
X, labels_true = make_blobs(
    n_samples=1000, centers=centers, cluster_std=0.29, random_state=0
)
X = StandardScaler().fit_transform(X)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(
    eps= 0.4,
    min_samples= 5
)

labels = dbscan.fit_predict(X)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X[:,0],
    X[:,1],
    c=labels,
    cmap="rainbow"
)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("DBSCAN Clustering")

plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(X, labels)
```
