### **When to use the algorithms of Boosting** 

- AdaBoost
    - You want a simple, highly interpretable ensemble baseline on a clean dataset with minimal outliers.

- Gradient Boost
    - You have medium-sized numerical data and want high precision without needing to tune complex code parameters.

- XGBoost
    - Your dataset is large, contains missing values, requires fast training speeds, or you want the absolute highest possible predictive accuracy.
