```python
import seaborn as sns
df = sns.load_dataset("tips")
df.head()
```

```python
df.time
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df["time"] = encoder.fit_transform(df["time"])
```

```python
df.time.unique()
```

```python
x = df.drop(labels=["time"], axis= 1)
y = df.time
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.pipeline import Pipeline    # And end-to-end runnig pipeline from pre-processing to model training
from sklearn.impute import SimpleImputer # I is used to handel missing vakues
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
categorical_col = ["sex", "smoker","day"]
numerical_col = ["total_bill", "tip", "size"]
```

```python
## numerical pipeline

num_pipeline = Pipeline(
    steps = [
        ("imputer", SimpleImputer(strategy= "mean")),
        ("scaler", StandardScaler())
    ]
)   

## categorical pipeline

cat_pipeline = Pipeline(
    steps= [
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("OneHotEncoder", OneHotEncoder())
    ]
)
```

```python
preprocessor = ColumnTransformer(
    [
        ("num_pipeline", num_pipeline, numerical_col),
        ("cat_pipeline", cat_pipeline, categorical_col)
    ]
)
```

```python
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
```

```python
## Model Automation
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

models = {
    "Random-Forest" : RandomForestClassifier(),
    "Decisiontree" : DecisionTreeClassifier()
}
```

```python
def model_evaluation(X_train, X_test, y_train, y_test, models):
    report = {}
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        model_score = accuracy_score(y_test, y_pred)
        report[name] = model_score
    return report
```

```python
model_evaluation(X_train, X_test, y_train, y_test, models)
```
